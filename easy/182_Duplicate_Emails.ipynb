```python
input_data = """| id | email   |
| -- | ------- |
| 1  | a@b.com |
| 2  | c@d.com |
| 3  | a@b.com |"""
expected_output = """| Email   |
| ------- |
| a@b.com |"""
```

```python
input_df = create_df_from_leetcode_input(input_data)
expected_output_df = create_df_from_leetcode_input(expected_output)
```

```python
output_df = input_df.groupBy("email")\
    .agg(F.count("*").alias("cnt"))\
        .filter('cnt > 1')\
            .select(F.col("email").alias("Email"))
```

```python
check_acceptance(output_df, expected_output_df)
```
