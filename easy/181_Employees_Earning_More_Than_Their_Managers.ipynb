```python
employees_input = """| id | name  | salary | managerId |
| -- | ----- | ------ | --------- |
| 1  | Joe   | 70000  | 3         |
| 2  | Henry | 80000  | 4         |
| 3  | Sam   | 60000  | null      |
| 4  | Max   | 90000  | null      |"""
expected_output = """| Employee |
| -------- |
| Joe      |"""
```

```python
employees_df = create_df_from_leetcode_input(employees_input)
expected_output_df = create_df_from_leetcode_input(expected_output)
```

```python
employees = employees_df.alias("employees")
managers = employees_df.alias("managers")
output_df = employees.join(managers, employees.managerId == managers.id, "inner").filter(employees.salary > managers.salary).select(employees.name.alias("Employee"))
```

```python
check_acceptance(output_df, expected_output_df)
```
