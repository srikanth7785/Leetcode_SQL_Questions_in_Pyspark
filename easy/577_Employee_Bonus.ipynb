```python
employee_input = """| empId | name   | supervisor | salary |
| ----- | ------ | ---------- | ------ |
| 3     | Brad   | null       | 4000   |
| 1     | John   | 3          | 1000   |
| 2     | Dan    | 3          | 2000   |
| 4     | Thomas | 3          | 4000   |"""
bonus_input = """| empId | bonus |
| ----- | ----- |
| 2     | 500   |
| 4     | 2000  |"""
expected_output = """| name | bonus |
| ---- | ----- |
| Brad | null  |
| John | null  |
| Dan  | 500   |"""
```

```python
employee_df = create_df_from_leetcode_input(employee_input)
bonus_df = create_df_from_leetcode_input(bonus_input)
expected_output_df = create_df_from_leetcode_input(expected_output)
```

```python
output_df = employee_df.join(bonus_df, "empId", "left")\
    .filter("bonus is null or bonus < 1000")\
        .select(
            "name",
            F.coalesce("bonus", F.lit("null")).alias("bonus")
            )
```
