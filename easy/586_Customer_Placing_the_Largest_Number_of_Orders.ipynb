```python
orders_df = create_df_from_leetcode_input(orders_input)
expected_output_df = create_df_from_leetcode_input(expected_output)
```

```python
output_df = orders_df.groupBy("customer_number").agg(F.count("*").alias("cnt")).orderBy(F.col("cnt").desc()).limit(1).select("customer_number")
```
