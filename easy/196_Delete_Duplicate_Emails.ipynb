```python
input_data = """| id | email            |
| -- | ---------------- |
| 1  | john@example.com |
| 2  | bob@example.com  |
| 3  | john@example.com |"""
expected_output = """| id | email            |
| -- | ---------------- |
| 1  | john@example.com |
| 2  | bob@example.com  |"""
```

```python
input_df = create_df_from_leetcode_input(input_data)
expected_output_df = create_df_from_leetcode_input(expected_output)
```

```python
output_df = input_df.groupBy("email").agg(F.min("id").alias("id")).select("id", "email")
```
