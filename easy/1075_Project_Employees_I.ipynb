```python
project_input = """| project_id | employee_id |
| ---------- | ----------- |
| 1          | 1           |
| 1          | 2           |
| 1          | 3           |
| 2          | 1           |
| 2          | 4           |"""
employee_input = """| employee_id | name   | experience_years |
| ----------- | ------ | ---------------- |
| 1           | Khaled | 3                |
| 2           | Ali    | 2                |
| 3           | John   | 1                |
| 4           | Doe    | 2                |"""
expected_output = """| project_id | average_years |
| ---------- | ------------- |
| 1          | 2             |
| 2          | 2.5           |"""
```

```python
project_df = create_df_from_leetcode_input(project_input)
employee_df = create_df_from_leetcode_input(employee_input)
expected_output_df = create_df_from_leetcode_input(expected_output)
```

```python
output_df = project_df.join(employee_df, "employee_id", "inner")\
    .groupBy("project_id")\
        .agg(
            F.round(
                F.avg(F.col("experience_years")),
                2
                ).alias("average_years")
            )
```
