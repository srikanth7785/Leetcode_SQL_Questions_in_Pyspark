```python
triangle_input = """| x  | y  | z  |
| -- | -- | -- |
| 13 | 15 | 30 |
| 10 | 20 | 15 |"""
expected_output = """| x  | y  | z  | triangle |
| -- | -- | -- | -------- |
| 13 | 15 | 30 | No       |
| 10 | 20 | 15 | Yes      |"""
```

```python
triangle_df = create_df_from_leetcode_input(triangle_input)
expected_output_df = create_df_from_leetcode_input(expected_output)
```

```python
def cast_all_single_letter_cols_to_int(input_df):
    return input_df.select([F.col(c).cast("int") if len(c) == 1 else F.col(c) for c in input_df.columns])
```

```python
triangle_df = cast_all_single_letter_cols_to_int(triangle_df)
expected_output_df = cast_all_single_letter_cols_to_int(expected_output_df)
```

```python
output_df = triangle_df.withColumn(
    "triangle",
    F.when(
        (F.col("x") + F.col("y") > F.col("z"))
        & (F.col("y") + F.col("z") > F.col("x"))
        & (F.col("x") + F.col("z") > F.col("y")),
        F.lit("Yes")
    ).otherwise(
        F.lit("No")
    )
)
```
