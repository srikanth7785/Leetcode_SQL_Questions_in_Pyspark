```python
product_input = """| product_id | product_name | unit_price |
| ---------- | ------------ | ---------- |
| 1          | S8           | 1000       |
| 2          | G4           | 800        |
| 3          | iPhone       | 1400       |"""
sales_input = """| seller_id | product_id | buyer_id | sale_date  | quantity | price |
| --------- | ---------- | -------- | ---------- | -------- | ----- |
| 1         | 1          | 1        | 2019-01-21 | 2        | 2000  |
| 1         | 2          | 2        | 2019-02-17 | 1        | 800   |
| 2         | 2          | 3        | 2019-06-02 | 1        | 800   |
| 3         | 3          | 4        | 2019-05-13 | 2        | 2800  |"""
expected_output = """| product_id | product_name |
| ---------- | ------------ |
| 1          | S8           |"""
```

```python
product_df = create_df_from_leetcode_input(product_input)
sales_df = create_df_from_leetcode_input(sales_input)
expected_output_df = create_df_from_leetcode_input(expected_output)
```

```python
output_df = product_df.join(sales_df, "product_id", "inner")\
    .groupBy("product_id").agg(
        F.min("sale_date").alias("min_sale_date"),
        F.max("sale_date").alias("max_sale_date")
        ).filter(
            (F.col("min_sale_date") >= '2019-01-01')
            & (F.col("max_sale_date") <= '2019-03-31')
        ).join(product_df, "product_id", "inner")\
            .select("product_id", "product_name")
output_df.display()
```
