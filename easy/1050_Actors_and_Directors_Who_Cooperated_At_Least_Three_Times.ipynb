```python
actor_director_input = """| actor_id | director_id | timestamp |
| -------- | ----------- | --------- |
| 1        | 1           | 0         |
| 1        | 1           | 1         |
| 1        | 1           | 2         |
| 1        | 2           | 3         |
| 1        | 2           | 4         |
| 2        | 1           | 5         |
| 2        | 1           | 6         |"""
expected_output = """| actor_id | director_id |
| -------- | ----------- |
| 1        | 1           |"""
```

```python
actor_director_df = create_df_from_leetcode_input(actor_director_input)
expected_output_df = create_df_from_leetcode_input(expected_output)
```

```python
output_df = actor_director_df.groupBy("actor_id","director_id")\
    .count()\
        .filter("count >= 3")\
            .select("actor_id", "director_id")
```
