```python
weather_input = """| id | recordDate | temperature |
| -- | ---------- | ----------- |
| 1  | 2015-01-01 | 10          |
| 2  | 2015-01-02 | 25          |
| 3  | 2015-01-03 | 20          |
| 4  | 2015-01-04 | 30          |"""
expected_output = """| Id |
| -- |
| 2  |
| 4  |"""
```

```python
weather_df = create_df_from_leetcode_input(weather_input)
expected_output_df = create_df_from_leetcode_input(expected_output)
```

```python
wind_spec = Window.orderBy("recordDate")
output_df = weather_df\
    .withColumn("yesterday_temperature", F.lag("temperature").over(wind_spec))\
        .filter("temperature > yesterday_temperature")\
            .select("id")
```
