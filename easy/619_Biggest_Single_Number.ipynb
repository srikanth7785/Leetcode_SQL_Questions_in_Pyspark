```python
my_numbers_input = """| num |
| --- |
| 8   |
| 8   |
| 3   |
| 3   |
| 1   |
| 4   |
| 5   |
| 6   |"""
expected_output = """| num |
| --- |
| 6   |"""
```

```python
my_numbers_input_df = create_df_from_leetcode_input(my_numbers_input)
expected_output_df = create_df_from_leetcode_input(expected_output)
```

```python
output_df = my_numbers_input_df.groupBy("num")\
    .count()\
        .filter("count = 1")\
            .orderBy(F.col("num").desc())\
                .limit(1)\
                    .select("num")
```
